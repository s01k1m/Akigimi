```python
import requests
naver_sports_url = "https://sports.news.naver.com/index.nhn"
naver_sports_main = requests.get(naver_sports_url)
```

```python
from bs4 import BeautifulSoup
naver_sports_main_list = BeautifulSoup(naver_sports_main.content,"html.parser",from_encoding=
'utf-8')
print(naver_sports_main_list)
```

```python
link_list = naver_sports_main_list.select(".vod_section >.vod_list >.vod_item")
for link in link_list:
    name = link.find(class_="text_area _sports_home_most_viewed_vod_title")
    print(name.text.strip())
```
